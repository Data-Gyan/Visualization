Matplotlib: Visualization with Python

Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python. Matplotlib makes easy things easy and hard things possible.

https://matplotlib.org

```python
import matplotlib.pyplot as plt
import numpy as np
# make sure the graphs are visile in jupyter notebook
%matplotlib inline
```

```python
areas = ['Advertisement', 'Sales', 'Inventory', 'Training']
allocation = [12, 34, 43, 11]
```

```python
plt.pie(allocation, labels=areas, autopct='%0.1f%%')
plt.show()
```

```python
X = [76.0, 76.6, 12.4, 34.23, 67.0, 43, 67, 35]
Y = [12.4, 63, 43, 88.89, 67, 3, 54.99, 4.23]
```

```python
pl = plt.scatter(X, Y)
plt.xlabel("Age")
plt.ylabel("Income")
plt.title("Income distribution vs Age")

plt.show()
```

```python
product_id = np.arange(1,9,1)
online_price = [12,32,43,43,24,65,34,63] 
offline_price = [34,22,33,54,20,34,73,61] 
localmart_price = [24,35,44,33,34,55,24,64] 
plt.plot(product_id, online_price, color='r', marker='D', markersize=8, label="online_price")
plt.plot(product_id, offline_price, color='b', marker='D', markersize=8, label="offline_price")
plt.plot(product_id, localmart_price, color='g', marker='D', markersize=8, label="localmart_price")

plt.xlabel("ProductID")
plt.ylabel("Online Price")
plt.title("Product vs Price")
#plt.legend(["online_price", "offline_price", "localmart_price"])
plt.legend()
plt.show()
```

```python
year = [2012, 2013, 2014, 2015, 2016, 2017, 2018]
price = [45034, 52000, 76983, 84343, 34084, 72343, 59999 ]
#plt.bar(year, price, color='g')
plt.barh(year, price, color='g')
plt.xlabel('year')
plt.ylabel('price/INR')
plt.title("price vs year comparison")
plt.show()
```
